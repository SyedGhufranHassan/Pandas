Import pandas

```python
import pandas as pd
```

Load Dataset

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

Handling Missing Values

Drop missing values unsig dropna()

When we apply dropna() on complete dataset it drop the all rows where he find any null value in data.

```python
df.isnull()
```

```python
df['Cabin'].dropna().shape
```

```python
df.shape
```

```python
df.dropna().shape
```

```python
df.dropna()
```

Now we apply how='all' then he drop those rows where he find all columns null for those rows.

```python
df.dropna(how='all')
```

```python
df.dropna(how='all').shape
```

By default in this function axis=0, but we can chage it to axis=1 and now he drop those columns where he find any null value.

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1).shape
#drop 2 columns
```

When we want to apply dropna on only  1 or 2 column

```python
df.dropna(subset=['Cabin']).shape
```

```python
df.dropna(subset=['Embarked']).shape
```

```python
df.dropna(subset=['Cabin','Embarked']).shape
```

Filling missing values 

Fill with fillna()

```python
df.fillna(0)
```

```python
df['Cabin'].fillna('Not',inplace=True)
```

```python
df
```

```python
df['Embarked'].fillna('A',inplace=True)
```

```python
df['Age'].fillna(25)
```

We also apply method=ffill, bfill

ffill=forward value

bfill=backward fill

```python
df['Age'].fillna(method='bfill')
```
